### 1. 품목 종류 분석

```python
test_arr_2018 = df_2018["품목 이름"].unique()
test_arr_2019 = df_2019["품목 이름"].unique()
test_arr_2020 = df_2020["품목 이름"].unique()
test_arr_2021 = df_2021["품목 이름"].unique()
test_arr_2022 = df_2022["품목 이름"].unique()
print(len(test_arr_2018))
print(len(test_arr_2019))
print(len(test_arr_2020))
print(len(test_arr_2021))
print(len(test_arr_2022))
```

```python
def split_and_set(seq):
  test_arr_2 = []
  for s in seq:
    ss = str(s).split("(")
    if ss[0] != "nan":
      test_arr_2.append(ss[0])
    test_arr_2 = list(set(test_arr_2))
  return(test_arr_2)
```

```python
result_2018 = split_and_set(test_arr_2018)
result_2019 = split_and_set(test_arr_2019)
result_2020 = split_and_set(test_arr_2020)
result_2021 = split_and_set(test_arr_2021)
result_2022 = split_and_set(test_arr_2022)

result_2018.sort()
result_2019.sort()
result_2020.sort()
result_2021.sort()
result_2022.sort()

print(len(result_2018), result_2018)
print(len(result_2019), result_2019)
print(len(result_2020), result_2020)
print(len(result_2021), result_2021)
print(len(result_2021), result_2022)
```

```python
def split_with_sauce(seq):
  result_list = []
  ss = seq.split("[양념]")[0][5:-1].split('| ')
  for s in ss:
    s = s.split(' ')[0]
    result_list.append(s)
  return(result_list)
```
