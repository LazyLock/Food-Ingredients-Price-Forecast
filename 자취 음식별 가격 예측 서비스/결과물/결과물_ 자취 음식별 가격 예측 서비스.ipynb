```python
## 다중회귀모델을 통해 예측한 2023.07의 음식 가격 

# 무 상추 양파 오이 호박
vege_result_202307 = [1821.3250082907598,
 1969.412720402662,
 3738.88632057077,
 1147.3009248652827,
 1122.8034052256126]

# 달걀, 닭고기, 돼지고기
meat_result_202307 = [6976.2124753, 6279.93558451, 17919.69973517]

# 사과, 배
fruit_result_202307 = [2279.23179003, 4239.47330194]

# 고등어, 오징어
fish_result_202307 = [4433.46595271, 5152.95626781]
```

```python
import numpy as np

def seq_round(seq):
  result_arr = []
  for i in seq:
    s = np.round(i,-1)
    s = str(s)[:-2] + '원'
    result_arr.append(s)
  return result_arr
```

```python
vege_202307 = seq_round(vege_result_202307)
meat_202307 = seq_round(meat_result_202307)
fruit_202307 = seq_round(fruit_result_202307)
fish_202307 = seq_round(fish_result_202307)
```
