```python
import pandas as pd
import numpy as np
```

```python
df_scom = pd.read_csv('/content/drive/MyDrive/고등어 지수.csv', encoding='euc-kr')
df_sqd=pd.read_csv('/content/drive/MyDrive/오징어 지수.csv', encoding='euc-kr')
df_apl=pd.read_csv('/content/drive/MyDrive/사과 지수.csv', encoding='euc-kr')
df_pr=pd.read_csv('/content/drive/MyDrive/배 지수.csv', encoding='euc-kr')
df_price=pd.read_csv('/content/drive/MyDrive/품목별 가격 데이터 최종.csv', encoding='euc-kr').transpose()
```

```python
df_price.columns=df_price.iloc[0]
```

```python
df_price=df_price.drop("Unnamed: 0")
```

```python
print(df_price.index)
```

```python
df_scom.set_index('Unnamed: 0', inplace=True)
df_apl.set_index('Unnamed: 0', inplace=True)
df_pr.set_index('Unnamed: 0', inplace=True)
df_sqd.set_index('Unnamed: 0', inplace=True)
print(df_scom.index)
print(df_sqd.index)
print(df_apl.index)
print(df_pr.index)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
y_scom = df_price.iloc[:, 10] 
y_sqd = df_price.iloc[:, 9]
y_apl = df_price.iloc[:, 11]
y_pr = df_price.iloc[:, 6]
```

고등어

```python
x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(df_scom, y_scom, test_size=0.2, random_state=13)

# 나누어진 데이터 세트 확인
print("훈련 세트 X:", x_train_1.shape)
print("타겟 세트 y:", y_train_1.shape)
print("테스트 세트 X:", x_test_1.shape)
print("타겟 세트 y:", y_test_1.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model_scom = LinearRegression()
```

```python
model_scom.fit(x_train_1, y_train_1)

y_pred_1 = model_scom.predict(x_test_1)
mse = mean_squared_error(y_test_1, y_pred_1)
r2 = r2_score(y_test_1, y_pred_1)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R^2):", r2)
```

```python
# 예측
index_data_2023_scom = [106.53116691913496, 103.72287970219153, 134.39700239653862, 94.95125326672229, 122.85438393793895] 
predicted_price_2023 = model_scom.predict([index_data_2023_scom])
print("2023년 예측 가격:", predicted_price_2023)
```
