```python
x_raddish = df_raddish[['소비자물가지수(단일)', '신선채소 (신선식품지수)', '냉동채소 (수입물가지수)', '총지수 (품목별소비자물가)',
       'Food Price Index', '식품']]
y_raddish = df_price["무"]

print(len(x_raddish))
print(len(y_raddish))
print(x_raddish.shape)
print(y_raddish.shape)
```

```python
## 무 가격 예측 모델

from sklearn.model_selection import train_test_split

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(x_raddish, y_raddish, test_size=0.2, random_state=42)

# 나누어진 데이터 세트 확인
print("훈련 세트 X:", x_train_1.shape)
print("타겟 세트 y:", y_train_1.shape)
print("테스트 세트 X:", x_test_1.shape)
print("타겟 세트 y:", y_test_1.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model_raddish = LinearRegression()
model_raddish.fit(x_train_1, y_train_1)

y_pred_1 = model_raddish.predict(x_test_1)
mse = mean_squared_error(y_test_1, y_pred_1)
r2 = r2_score(y_test_1, y_pred_1)

print("Mean Squared Error (MSE):", mse)
print("R-squared (R^2):", r2)
```
