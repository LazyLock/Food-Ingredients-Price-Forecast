```python
import pandas as pd

file_path = "/content/drive/MyDrive/23-1 ITS 데이터 분석 스터디 : 프로젝트 1팀/데이터 아카이브/추가 데이터/제경/물가지수/"
file_name = "품목별소비자물가지수_201801~202306.csv"

df = pd.read_csv(file_path + file_name, encoding='cp949')
print(len(df))
```

```python
def is_in_mtr_arr(seq):
  mtr_arr =['무', '갈치', '고등어', '닭고기', '양파', '배추', '냉동참조기', '오징어', '조기', '애호박', '사과', '동태', '호박', '배', '쇠고기', '달걀', '상추', '돼지고기', '오이', '명태', '삼겹살', '쌀']
  if seq in mtr_arr:
    return True
  else:
    return False
```

```python
mtr_arr =['무', '갈치', '고등어', '닭고기', '양파', '배추', '냉동참조기', '오징어', '조기', '애호박', '사과', '동태', '호박', '배', '쇠고기', '달걀', '상추', '돼지고기', '오이', '명태', '삼겹살', '쌀']

df_new = df["품목별"].apply(lambda x: is_in_mtr_arr(x))
len(df_new)
```

```python
df = df[df_new]
```

```python
df_new = df.reset_index()
df_new = df_new.drop(["index"], axis=1)

df_new.info()
```

```python
mtr_from_df = df_new["품목별"]
mtr_from_df = list(mtr_from_df)

result = []

for s in mtr_arr:
  if s in mtr_from_df:
    continue
  else:
    result.append(s)

result. 
```

```python
## 위 5개 항목을 제외하고 2018.01 ~ 2023.05까지의 물가지수 데이터는 잘 보존되어 있네요
## 칼럼명만 YYYY_MM 형태로 통일할게요

def change_date_format(s):
  s = s[:-2].split('.')
  s = '-'.join(s)
  return s

df_date_columns = list(map(change_date_format, df_new.columns[4:]))
df_new_date_columns = [*df_new.columns[:4], *df_date_columns]

df_new.columns = df_new_date_columns
df_new.head()
```
