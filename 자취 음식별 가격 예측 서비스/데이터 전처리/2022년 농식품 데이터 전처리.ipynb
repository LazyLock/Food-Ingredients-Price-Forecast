```python
def split_with_semicolon(seq):
  result_arr = []
  for s in seq:
    split_arr = str(s).split(";")
    result_arr = [*result_arr, *split_arr]
  return result_arr

test11 = df_2022.iloc[27].dropna()
print(list(test11))
print(split_with_semicolon(test11))
print(len(split_with_semicolon(test11)))
```

```python
def change_type(seq):
  result_arr = []
  i = 0
  while i < len(seq):
    if seq[i][0] == seq[i][-1]:
      result_arr.append(eval(seq[i]))
      i += 1
    elif seq[i][0] ==  '"':
      if seq[i+1][-1] != '"':
        ss = seq[i] + seq[i+1] + seq[i+2]  
        result_arr.append(eval(ss))
        i += 3
      else:
        ss = seq[i] + seq[i+1]
        result_arr.append(eval(ss))
        i += 2
    elif seq[i][-1] != '"':
      result_arr.append(seq[i])
      i += 1
  return result_arr

test11 = list(df_2022.iloc[27].dropna())
print(change_type(split_with_semicolon((test11))))
print(len(change_type(split_with_semicolon((test11)))))

# print(len(change_type(split_with_semicolon(test11))))
```

```python
i=0
while i < 1000:
    test_arr_2022 = list(df_2022.iloc[i].dropna())
    len_test = len(change_type(split_with_semicolon(test_arr_2022)))
    if len_test != 14:
      print(i)
    i += 1
```

```python
import numpy as np

df_test_test = df_2022.values
test123 = df_test_test[1]

test123 = [ s for s in test123 if type(s) != float ]
test123

print(split_with_semicolon(test123))
print(change_type(split_with_semicolon((test123))))
print(len(change_type(split_with_semicolon((test123)))))
```

```python
i = 0
for_df_2022 = []

for seq in df_2022.values:
  try:
    new_seq = [ s for s in seq if type(s) != float ]
    new_seq = change_type(split_with_semicolon((new_seq)))
    for_df_2022.append(new_seq)
  except SyntaxError:
    pass
  except IndexError:
    pass

print(len(for_df_2022))
print(for_df_2022[:5])
```
