```python
def split_with_sauce(seq):
  result_list = []
  ss = seq.split("[양념]")[0][5:-1].split('| ')
  for s in ss:
    s = s.split(' ')[0]
    result_list.append(s)
  return(result_list)
```

```python
new_mtr_arr = []

for s in mtr_arr:
  new_mtr_arr.append(split_with_sauce(s))

new_mtr_arr[:10]
print(len(new_mtr_arr))
```

```python
arr_item_all = [*arr_2018, *arr_2019, *arr_2020, *arr_2021, *arr_2022 ]
arr_item_all = list(set(arr_item_all))
print(len(arr_item_all))
print(arr_item_all)
```

```python
protein = ['돼지고기', '닭고기', '삼겹살', '쇠고기', '달걀']
fish = ['동태', '고등어', '조기', '냉동참조기', '갈치', '오징어','명태']
fruit = ['배','사과']
vege = ['쌀','오이','호박','애호박','무','양파','배추','양파', '상추']

prot_dic = {}
fish_dic = {}
frut_dic = {}
vege_dic = {}

for s in protein:
  prot_dic[s] = 2

for s in fish:
  fish_dic[s] = 2

for s in fruit:
  frut_dic[s] = 1

for s in vege:
  vege_dic[s] = 1

# print(prot_dic)
# print(fish_dic)
# print(frut_dic)
# print(vege_dic)

all_dic = {}
all_dic.update(fish_dic)
all_dic.update(prot_dic)
all_dic.update(frut_dic)
all_dic.update(vege_dic)
print(all_dic)
```

```python
cnt = 0
mtr_result_arr = []
mtr_result_idx_arr = []

for idx, seq1 in enumerate(new_mtr_arr):
  for s in seq1:
    if s in arr_item_all:
      cnt += all_dic[s]
  if cnt >= 4:
    mtr_result_arr.append(seq1)
    mtr_result_idx_arr.append(idx)
  cnt = 0

print(mtr_result_arr)
print(mtr_result_idx_arr)
print(len(mtr_result_arr))
print(len(mtr_result_idx_arr))
```
