```python
## 박스 플롯 기준 분위수별 값 컷팅 (이상치 제거)

def set_quantiles(seq):
  quantiles = np.percentile(seq, [25, 50, 75])
  IQR = quantiles[2] - quantiles[0]
  max_seq = quantiles[2] + (1.5 * IQR)
  min_seq = quantiles[0] - (1.5 * IQR)
  return [min_seq, max_seq]

def del_outlier(i, min_seq, max_seq):
  if (i >= min_seq) & (i <= max_seq):
    return True
  else:
    return False

arr_apple_price = df_2022_apple["가격(원)"]
[min_apple, max_apple] = set_quantiles(arr_apple_price)
df_2022_apple = df_2022_apple[df_2022_apple["가격(원)"].apply(lambda x: del_outlier(x, min_apple, max_apple))]

# print(min(df_2022_apple["가격(원)"]))
# print(max(df_2022_apple["가격(원)"]))

df_2022_apple.groupby("년도-월")["가격(원)"].mean()
```
